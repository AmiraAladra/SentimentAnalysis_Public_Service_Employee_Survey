```python
from kmodes.kprototypes import KPrototypes
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Separate categorical & numerical columns
categorical_idx = [0,1,16,17,18,19]  # Columns containing categorical data
```

```python
# Identify numerical columns (all except categorical_idx)
num_cols = [col for col in df.columns if col not in categorical_idx]
```

```python
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')  # Convert any non-numeric values to NaN
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())  # Fill NaN values with the mean (or choose another strategy)
```

```python
# **Step 2: Check for NaN after handling**
if df[num_cols].isnull().any().any():
    print("There are still NaN values in the dataset.")
else:
    print("No NaN values left in the numerical columns.")
```

```python
# Standardize numerical features
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python

# Convert categorical columns to numerical labels for silhouette scoring
df_encoded = df.copy()
label_encoders = {}
for col in categorical_idx:
    le = LabelEncoder()
    df_encoded.iloc[:, col] = le.fit_transform(df.iloc[:, col])
    label_encoders[col] = le  # Store encoders for reference
```

```python
# Define the range of K values to test
K_range = range(2, 10)  # Trying different numbers of clusters
costs = []
sil_scores = []

# Iterate over possible values of K
for k in K_range:
    kproto = KPrototypes(n_clusters=k, init='Cao', n_init=5, gamma=5, verbose=1)
    clusters = kproto.fit_predict(df, categorical=categorical_idx)

    costs.append(kproto.cost_)  # Store cost for Elbow Method

    # Compute silhouette score using only numerical columns
    if k > 1:
        sil_scores.append(silhouette_score(df_encoded.iloc[:, num_cols], clusters))
    else:
        sil_scores.append(0)  # Silhouette score is invalid for
```

```python
# **Step 1: Find the best K using the Elbow Method**
plt.figure(figsize=(10, 5))
plt.plot(K_range, costs, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Cost')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# records where ANSWER3 or ANSWER4 or ANSWER5 or ANSWER6 or ANSWER7 = '9999'
df[(df['ANSWER3'] == '9999') | (df['ANSWER4'] == '9999') |(df['ANSWER5'] == '9999') |(df['ANSWER5'] == '9999') |(df['ANSWER7'] == '9999')]
```

```python
# Replace and value in the dataset that equal to 'X' with NaN
df = df.replace('X', pd.NA)
```

```python
df.info()
```

```python
# Change the data type of BYCOND to object
df['BYCOND'] = df['BYCOND'].astype(object)
```

```python
# Change the data type of DEMCODE to int
df['DEMCODE'] = df['DEMCODE'].astype('Int64')
```

```python
answers_cols = ['ANSWER1', 'ANSWER2', 'ANSWER3', 'ANSWER4', 'ANSWER5', 'ANSWER6', 'ANSWER7']
```

```python
df[answers_cols] = df[answers_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Change the data type of answers columns  to integer
df[answers_cols] = df[answers_cols].astype('Int64')
```

```python
# Select the desired columns using a list
cols_to_convert = ['MOST_POSITIVE_OR_LEAST_NEGATIVE', 'NEUTRAL_OR_MIDDLE_CATEGORY', 'MOST_NEGATIVE_OR_LEAST_POSITIVE','AGREE','SCORE5','SCORE100','ANSCOUNT']
# Apply pd.to_numeric to the selected columns
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
# Change the data type of columns ANSWER1 and ANSWER2 to integer
df[cols_to_convert] = df[cols_to_convert].astype(pd.Int64Dtype(),errors='ignore')
```

```python
answer_cols = ["ANSWER1", "ANSWER2", "ANSWER3", "ANSWER4", "ANSWER5", "ANSWER6"]
df[answer_cols] = df[answer_cols].apply(pd.to_numeric, errors='coerce')
```

```python

# CASE 1: If ANSWER3 is null
df.loc[df['ANSWER3'].isna(), 'MOST_POSITIVE_OR_LEAST_NEGATIVE'] = df['ANSWER1']
df.loc[df['ANSWER3'].isna(), 'MOST_NEGATIVE_OR_LEAST_POSITIVE'] = df['ANSWER2']

# CASE 2: If ANSWER3 is not null and ANSWER4 is null
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].isna(), 'MOST_POSITIVE_OR_LEAST_NEGATIVE'] = df['ANSWER1']
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].isna(), 'NEUTRAL_OR_MIDDLE_CATEGORY'] = df['ANSWER2']
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].isna(), 'MOST_NEGATIVE_OR_LEAST_POSITIVE'] = df['ANSWER3']

# CASE 3: If ANSWER3 is not null, ANSWER4 is not null, and ANSWER5 is null
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].notna() & df['ANSWER5'].isna(), 'MOST_POSITIVE_OR_LEAST_NEGATIVE'] = df['ANSWER1'] + df['ANSWER2']
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].notna() & df['ANSWER5'].isna(), 'NEUTRAL_OR_MIDDLE_CATEGORY'] = df['ANSWER3']
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].notna() & df['ANSWER5'].isna(), 'MOST_NEGATIVE_OR_LEAST_POSITIVE'] = df['ANSWER4']

# CASE 4: If ANSWER3, ANSWER4, ANSWER5 are not null, and ANSWER6 is null
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].notna() & df['ANSWER5'].notna() & df['ANSWER6'].isna(), 'MOST_POSITIVE_OR_LEAST_NEGATIVE'] = df['ANSWER1'] + df['ANSWER2']
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].notna() & df['ANSWER5'].notna() & df['ANSWER6'].isna(), 'NEUTRAL_OR_MIDDLE_CATEGORY'] = df['ANSWER3']
df.loc[df['ANSWER3'].notna() & df['ANSWER4'].notna() & df['ANSWER5'].notna() & df['ANSWER6'].isna(), 'MOST_NEGATIVE_OR_LEAST_POSITIVE'] = df['ANSWER4']
```
